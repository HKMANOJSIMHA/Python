```python
import random
def declare(inp):
    split=inp.split()
    output=''
    if('declare'==split[0].lower() and split[1].lower()=='number'):
        output='int '
        for i in range(2,len(split)):
            if(i==len(split)-1):
                 output=output+split[i]
            else:
                output=output+split[i]+','
        output=output+';'
        return output
    if('declare'==split[0].lower() and split[1].lower()=='string'):
        output='char '
        for i in range(2,len(split)):
            if(i==len(split)-1):
                output=output+split[i]+'['+str(random.randint(1,20))+']'
            
            else:
                output=output+split[i]+'['+str(random.randint(1,20))+']'+','
        output=output+';'
        return output
    if('declare'==split[0].lower() and split[1].lower()=='decimal'):
        output='float '
        for i in range(2,len(split)):
            if(i==len(split)-1):
                 output=output+split[i]
            else:
                output=output+split[i]+','
        output=output+';'
        return output
    if('declare'==split[0].lower() and split[1].lower()=='character'and split[2].lower()!='array'):
        output='char '
        for i in range(2,len(split)):
            if(i==len(split)-1):
                 output=output+split[i]
            else:
                output=output+split[i]+','
        output=output+';'
        return output
    if(split[1].lower()=='integer' and split[2]=='array'):
        output='int '
        output=output+split[3]+'['+split[5]+']'
        output=output+';'
        return output
        
    if(split[1].lower()=='floating' and split[2]=='array'):
        output='float '
        output=output+split[3]+'['+split[5]+']'
        output=output+';'
        return output
        
    if(split[1].lower()=='character' and split[2]=='array'):
        output='char '
        output=output+split[3]+'['+split[5]+']'
        output=output+';'
        return output
```

```python
def initialize(inp):
    split=inp.split()
    output=''
    if(split[2]=='of'):
        output=output+split[1]+'['+split[3]+']='+split[5]
        return output+';'
    
    elif(split[1]!='array'):
        output=output+split[1]+'='+split[3]
        return output+';'
    
    else:
        output=split[2]+'[]={ '
        for i in range(4,len(split)):
            if(i==len(split)-1):
                output=output+split[i]+' }'
            else:
                 output=output+split[i]+' '
        return output+';'
           
            
```

```python
def expression(inp):
    output=''
    split=inp.split()
    if(split[0]=='add'):
        output=split[1]
        for i in range(2,len(split)):
            if(split[i].lower()!='store'):
                output=output+'+'+split[i]
            else:
                break
        return split[len(split)-1]+'='+output+';'
    if(split[0]=='subtract'):
        output=split[1]
        for i in range(2,len(split)):
            if(split[i].lower()!='store'):
                output=output+'-'+split[i]
            else:
                break
        return split[len(split)-1]+'='+output+';'
    if(split[0]=='multiply'):
        output=split[1]
        for i in range(2,len(split)):
            if(split[i].lower()!='store'):
                output=output+'*'+split[i]
            else:
                break
        return split[len(split)-1]+'='+output+';'
    if(split[0]=='divide'):
        output=split[1]
        for i in range(2,len(split)):
            if(split[i].lower()!='store'):
                output=output+'/'+split[i]
            else:
                break
        return split[len(split)-1]+'='+output+';'
      
        
```

```python
def printSt(inp):
    split=inp.split()
    output=''
    if(split[1]=='numbers'):
            output='printf("'
            for i in range(2,len(split)):
                output=output+'%d'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+','+split[i]
            return output+');'
    elif(split[1]=='string'):
            output='printf("'
            for i in range(2,len(split)):
                output=output+'%s'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+','+split[i]
            return output+');'
    elif(split[1]=='characters'):
            output='printf("'
            for i in range(2,len(split)):
                output=output+'%c'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+','+split[i]
            return output+');'
    else:
        output='printf("'
        for i in range(1,len(split)):
                output=output+' '+split[i]
        return output+');'
```

```python
def scanSt(inp):
    split=inp.split()
    output=''
    if(split[1]=='numbers'):
            output='scanf("'
            for i in range(2,len(split)):
                output=output+'%d'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+',&'+split[i]
            return output+');'
    if(split[1]=='string'):
            output='scanf("'
            for i in range(2,len(split)):
                output=output+'%s'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+','+split[i]
            return output+');'
    if(split[1]=='characters'):
            output='scanf("'
            for i in range(2,len(split)):
                output=output+'%c'
            output=output+'"'
            for i in range(2,len(split)):
                output=output+',&'+split[i]
            return output+');'
            
```

```python
def ifSt(inp):
    split=inp.split()
    cond=getCond(split)
    return 'if('+cond+'){'
```

```python
def getCond(split):
    output=''
    if(split[2]=='greater'):
            return split[1]+'>'+split[4]
    if(split[2]=='less'):
            return split[1]+'<'+split[4]
    if(split[2]=='equal'):
            return split[1]+'=='+split[3]
    if(split[2]=='not'):
            return split[1]+'!='+split[5]
   
    
    
```

```python
def elseif(inp):
    split=inp.split()
    cond=getCond(split)
    return 'else if('+cond+'){'
```

```python
def whileSt(inp):
    split=inp.split()
    cond=getCond(split)
    return 'while('+cond+')'
```

```python
def inc_dec(inp):
    split=inp.split()
    if(split[0]=='incriment'):
        return split[1]+'++;'
    if(split[0]=='decriment'):
        return split[1]+'--;'
```

```python
def forSt(inp):
    split=inp.split()
    if(int(split[3])<int(split[5])):
        return 'for('+split[1]+'='+split[3]+';'+split[1]+'<'+split[5]+';'+split[1]+'++){'
    if(int(split[3])>int(split[5])):
        return 'for('+split[1]+'='+split[3]+';'+split[1]+'>'+split[5]+';'+split[1]+'--){'
```

```python
def enddo(inp):
    split=inp.split()
    cond=getCond(split)
    return 'while('+cond+');'
```

```python
def define(inp):
    split=inp.split()
    return '#define '+split[1]+' '+split[3]
```

```python
inp=[]
output='#include<stdio.h>\n#include<stdlib.h>\n'
print('Enter the algorithm')
String=''
while String!='end':
    String=input()
    inp.append(String)
for string in inp:
    split=string.split()
    if(split[0]=='define'):
        output=output+'\n'+define(string)
    if(split[0]=='start'):
        output=output+'\n'+'main(){'
    if(split[0]=='declare'):
        output=output+'\n'+declare(string)
    if(split[0]=='initialize'):
        output=output+'\n'+initialize(string)
    if(split[0]=='print'):
        output=output+'\n'+printSt(string)
    if(split[0]=='scan'):
        output=output+'\n'+scanSt(string)
    if(split[0]=='add' or split[0]=='subtract' or split[0]=='multiply' or split[0]=='divide'):
        output=output+'\n'+expression(string)
    if(split[0]=='incriment'or split[0]=='decriment'):
        output=output+'\n'+inc_dec(string)
    if(split[0]=='if'):
         output=output+'\n'+ifSt(string)
    if(split[0]=='else'):
        output=output+'\nelse{'
    if(split[0]=='elseif'):
        output=output+'\n'+elseif(string)
    if(split[0]=='while'):
        output=output+'\n'+whileSt(string)
    if(split[0]=='for'):
        output=output+'\n'+forSt(string) 
    if(split[0]=='do'):
        output=output+'\ndo{'
    if(split[0]=='until'):
        output=output+'\n}'+enddo(String)
    if(split[0]=='endif'):
        output=output+'\n}'
    if(split[0]=='endelse'):
        output=output+'\n}'
    if(split[0]=='endelif'):
        output=output+'\n}'
    if(split[0]=='endwhile'):
        output=output+'\n}'
    if(split[0]=='endfor'):
         output=output+'\n}'
    if(split[0]=='end'):
        output=output+'\n}'
print('\n'+output+'\n')
```
